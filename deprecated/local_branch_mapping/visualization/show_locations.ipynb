```python
def get_country_shapefile(country, admin_level=0):
    """
    get shapefile of given country
    """
    country_shapefile = {
            'Guatemala': 'gtm_admbnda_adm'+str(admin_level)+'_ocha_conred_20190207.shp',
            'Lebanon': 'OCHA/lbn_admbnda_adm'+str(admin_level)+'_cdr_20160822.shp',
            'Netherlands': 'NLD_adm'+str(admin_level)+'.shp',
            'Malawi': 'adm'+str(admin_level)+'.shp'
            }
    return str('shapefiles/'+country+'/'+country_shapefile[country])
```
